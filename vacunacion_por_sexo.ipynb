Análisis vacunación COVID-19 por sexo para primera y segunda dosis - Total País - Argentina 2021

```python
#import librerias necesarias

import numpy as np 
import pandas as pd
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from zipfile import ZipFile
from urllib.request import urlopen   
import pandas as pd
import os
```

```python
#Analisis de casos registrados Covid19 Argentina 
myUrl = 'https://sisa.msal.gov.ar/datos/descargas/covid-19/files/datos_nomivac_covid19.zip'
# read the zip file as a pandas dataframe
data = pd.read_csv(myUrl)   # pandas version 0.18.1 takes zip files       
# Creamos nuestro DF en base al Dasaset 
covid = pd.DataFrame(data)
#Verificar el tipo de datos del data frame obtenido
#covid.dtypes
#Contador de vacunados segun sexo (F o M) para la 1ra dosis
fem1 = covid[(covid['sexo'] == 'F') & (covid['orden_dosis'] == 1)].shape[0]
mas1 = covid[(covid['sexo'] == 'M') & (covid['orden_dosis'] == 1)].shape[0]
#Total de vacunados con 1ra dosis
total1 = fem1+mas1
print("Vacunación primera dosis (femeninos): " ,fem1)
print("Vacunación primera dosis (masculinos): ",mas1)
print("Total vacunados primera dosis: ",total1)
#Contador de vacunados segun sexo (F o M) para la 2da dosis
fem2 = covid[(covid['sexo'] == 'F') & (covid['orden_dosis'] == 2)].shape[0]
mas2 = covid[(covid['sexo'] == 'M') & (covid['orden_dosis'] == 2)].shape[0]
#Total de vacunados con 2da dosis
total2 = fem2+mas2
print("Vacunación segunda dosis (femeninos): " ,fem2)
print("Vacunación segunda dosis (masculinos): ",mas2)
print("Total vacunados segunda dosis: ",total2)

#totales = np.array([total1,total2])

# 1,2 denotes 1 row, 2 columns - if you want to stack vertically, it would be 2,1
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,10)) #ax1,ax2 refer to your two pies
#labels
labels = ['Femenino','Masculino']
#colors
colors = ['#AB47BC','#26A69A']
#array con valores de vacunados con primera dosis segun sexo
prim_dosis = np.array([fem1,mas1])

#grafico con % de vacunados segun sexo (1ra dosis)
patches, texts, autotexts = ax1.pie(prim_dosis, colors = colors, labels=labels, autopct='%1.1f%%', startangle=180, shadow=True)
for text in texts:
    text.set_color('grey')
for autotext in autotexts:
    autotext.set_color('white')
#titulo del grafico
ax1.set_title('Primera Dosis', y=0.8)

#labels
labels = ['Femenino','Masculino']
#colors
colors = ['#AB47BC','#26A69A']
#array con valores de vacunados con segunda dosis segun sexo
seg_dosis = np.array([fem2,mas2])

#grafico con % de vacunados segun sexo (2da dosis)
patches, texts, autotexts = ax2.pie(seg_dosis, colors = colors, labels=labels, autopct='%1.1f%%', startangle=180, shadow=True)
for text in texts:
    text.set_color('grey')
for autotext in autotexts:
    autotext.set_color('white')
ax2.set_title('Segunda Dosis',y=0.8)

#titulo del grafico general
fig.suptitle("% Vacunación por sexo", y=0.85)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
ax2.axis('equal')
plt.tight_layout()
plt.show()

```
